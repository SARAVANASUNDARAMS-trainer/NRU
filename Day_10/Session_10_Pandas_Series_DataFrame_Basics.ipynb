## 1. What is Pandas?

Pandas is an open-source Python library used for data manipulation and analysis. It is built on top of **NumPy** and is designed to work with **structured/tabular data** (similar to an Excel sheet or SQL table).

**Why use pandas?**
- Labeled data — rows and columns have names
- Fast vectorized operations
- Handles missing data gracefully
- Rich set of tools for reading, cleaning, transforming, and analyzing data


```python
import pandas as pd
import numpy as np
```

## 2. Two Core Data Structures

| Structure | Dimensionality | Analogy |
|---|---|---|
| **Series** | 1-D | Single column of Excel / a list with labels |
| **DataFrame** | 2-D | Full Excel sheet / SQL table |

> A DataFrame is essentially a collection of Series sharing the same index.


## 3. Creating a Series

```python
# From a list
s1 = pd.Series([10, 20, 30, 40])
print(s1)
```

```python
# From a list with custom index
s2 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(s2)
```

```python
# From a dictionary (keys become index)
s3 = pd.Series({'math': 85, 'science': 90, 'english': 78})
print(s3)
```

```python
# From a scalar value (broadcast)
s4 = pd.Series(5, index=[0, 1, 2, 3])
print(s4)
```

**Key observations:**
- Default index starts at 0 (like array position)
- Each Series has a **dtype** (int64, float64, object, etc.)
- Series = values + index (unlike a plain Python list)


## 4. Series Attributes

```python
print(s2.values)   # underlying numpy array
print(s2.index)    # index object
print(s2.dtype)    # data type of elements
print(s2.shape)    # tuple showing size
print(s2.size)     # number of elements
print(s2.name)     # optional name of the series
```

```python
import pandas as pd

ss11 = pd.Series([10, 20, 30], name="Marks")

print(ss11.name)
```

```python
s2
```

## 5. Indexing and Selecting from a Series

```python
print(s2['a'])       # label-based access
print(s2.loc['b'])   # explicit label-based access
print(s2.iloc[1])    # explicit position-based access
```

```python
print(s2['a':'c'])   # slicing by label (inclusive of end)
print(s2[0:2])       # slicing by position (exclusive of end)
```

**Important note:**
- `.loc[]` → label-based
- `.iloc[]` → integer position-based

This distinction is important and applies to DataFrames as well.


## 6. Basic Series Operations

```python
print(s2 + 5)          # broadcasting
print(s2 * 2)
print(s2[s2 > 15])     # boolean filtering
print(s2.sum(), s2.mean(), s2.max(), s2.min())
```

## 7. What is a DataFrame?

A DataFrame is a 2-dimensional labeled data structure with columns that can hold different data types. It can be thought of as a dictionary of Series objects that share a common index.


## 8. Creating a DataFrame

```python
# From a dictionary of lists
data = {
    'Name': ['Asha', 'Ravi', 'Meena', 'Karan'],
    'Age': [23, 25, 22, 24],
    'City': ['Chennai', 'Delhi', 'Mumbai', 'Pune'],
    'Marks': [85, 78, 92, 88]
}
df = pd.DataFrame(data)
df
```

```python
# From a list of dictionaries
records = [
    {'Name': 'Asha', 'Age': 23},
    {'Name': 'Ravi', 'Age': 25}
]
df2 = pd.DataFrame(records)
df2
```

```python
# From a list of lists with explicit columns
df3 = pd.DataFrame(
    [[1, 'A'], [2, 'B']],
    columns=['ID', 'Grade']
)
df3
```

```python
# From a NumPy array
df4 = pd.DataFrame(np.arange(1, 13).reshape(3, 4), columns=['W', 'X', 'Y', 'Z'])
df4
```

## 11. Selecting Columns

## 12. Selecting Rows

```python
df.loc[0]           # row by label
```

```python
df.iloc[0]           # row by position
```

```python
df.loc[0:2]          # label-based slicing (inclusive)
```

```python
df.iloc[0:2]         # position-based slicing (exclusive)
```

## 13. Adding and Modifying Columns

```python
df['Passed'] = df['Marks'] >= 80          # new boolean column
df
```

```python
df['Marks_Percent'] = df['Marks'] / 100 * 100
df
```

```python
df['Age'] = df['Age'] + 1                 # modify existing column
df
```

```python
df.rename(columns={'Marks': 'Score'}, inplace=True)
df
```

## 14. Dropping Rows/Columns

```python
df.drop('Passed', axis=1)              # drop column (axis=1)
```

```python
df.drop(0, axis=0)                      # drop row (axis=0)
```

```python
df.drop(columns=['City'], inplace=True) # inplace modification
df
```

## Key Terms Glossary

| Term | Meaning |
|---|---|
| Series | 1-dimensional labeled array |
| DataFrame | 2-dimensional labeled tabular structure |
| Index | Labels for rows |
| dtype | Data type of column/Series values |
| loc | Label-based indexer |
| iloc | Integer position-based indexer |
| axis=0 | Refers to rows |
| axis=1 | Refers to columns |


## Common Mistakes to Avoid

- Confusing `.loc[0:2]` (inclusive) with `.iloc[0:2]` (exclusive)
- Forgetting `axis=1` when dropping columns
- Using `df['col1', 'col2']` instead of `df[['col1', 'col2']]` for multiple columns
- Not using `inplace=True` or reassigning the result, then wondering why changes don't persist
