## 3. Reading CSV Files — The Basics


```python
df = pd.read_csv('students.csv')
df
```

### Commonly used `read_csv()` parameters

| Parameter | Purpose |
|---|---|
| `filepath_or_buffer` | Path to the CSV file (or a URL) |
| `sep` / `delimiter` | Character used to separate values (default `,`) |
| `header` | Row number to use as column names (default `0`) |
| `names` | Manually provide column names |
| `index_col` | Column to use as the row index |
| `usecols` | Load only specific columns |
| `nrows` | Load only a specific number of rows |
| `skiprows` | Skip a specific number of rows at the start |
| `dtype` | Force specific data types for columns |
| `na_values` | Additional strings to treat as missing values |
| `encoding` | Character encoding of the file (e.g. `'utf-8'`, `'latin1'`) |

## 7. Selecting Rows

Two primary indexers:
- `.loc[]` — label-based selection (uses row/column **names**)
- `.iloc[]` — position-based selection (uses integer **positions**, like array indices)


```python
df.loc[0]              # row by label
```

```python
df.iloc[0]             # row by position
```

```python
df.loc[0:3]            # label-based slicing (inclusive of end -- rows 0,1,2,3)
```

```python
df.iloc[0:3]           # position-based slicing (exclusive of end -- rows 0,1,2)
```

```python
df.loc[[1, 3, 5]]      # select specific non-contiguous rows by label
```

```python
df.iloc[[1, 3, 5]]     # select specific non-contiguous rows by position
```

## 8. Selecting Rows and Columns Together


```python
df.loc[1, 'Name']                     # single cell by label
```

```python
df.iloc[1, 1]                          # single cell by position
```

```python
df.loc[0:3, ['Name', 'Department']]    # subset of rows and columns by label
```

```python
df.iloc[0:3, 0:3]                      # subset of rows and columns by position
```

```python
df.loc[:, 'Name':'Age']                # all rows, range of columns by label
```

## 10. Conditional (Boolean) Row Selection

Rows can be filtered based on conditions applied to one or more columns. (A deeper treatment of filtering, sorting, and grouping is in the next session — this section builds the foundation.)


```python
df[df['CGPA'] > 8]                             # single condition
```

```python
df[df['Department'] == 'CS']                   # equality condition
```

```python
df[(df['CGPA'] > 7.5) & (df['Department'] == 'CS')]    # AND condition
```

```python
df[(df['Department'] == 'CS') | (df['Department'] == 'ME')]   # OR condition
```

```python
df[df['Department'].isin(['CS', 'ME'])]        # isin() -- cleaner alternative to multiple OR conditions
```

```python
df[df['Age'].between(21, 22)]                  # between() -- range condition
```

```python
df[~(df['Department'] == 'CS')]                # NOT condition using ~
```

## Key Terms Glossary

| Term | Meaning |
|---|---|
| `read_csv()` | Function to load a CSV file into a DataFrame |
| `to_csv()` | Function to save a DataFrame as a CSV file |
| `sep` | Delimiter character used in the file |
| `header` | Row used as column names |
| `index_col` | Column used as row index |
| `usecols` / `nrows` / `skiprows` | Control which part of the file is loaded |
| `head()` / `tail()` / `sample()` | View rows of a DataFrame |
| `unique()` / `nunique()` / `value_counts()` | Explore distinct values in a column |
| `loc` | Label-based indexer |
| `iloc` | Integer position-based indexer |
| `at` / `iat` | Fast scalar (single value) access |
| `isin()` | Check membership in a list of values |
| `between()` | Check if values fall within a range |
| Boolean selection | Filtering rows using a condition |


## Common Mistakes to Avoid

- Forgetting `index=False` when saving to CSV (creates an unwanted extra `Unnamed: 0` column)
- Confusing `.loc[0:3]` (inclusive of end) with `.iloc[0:3]` (exclusive of end)
- Using `df['col1', 'col2']` instead of `df[['col1', 'col2']]` for multiple columns
- Forgetting the right `sep` value when a file uses a delimiter other than a comma
- Not checking `.info()` / `.isnull().sum()` after reading a CSV, missing wrong data types or missing values
- Using `header=None` without also providing `names=`, which leaves columns labeled `0, 1, 2...`
