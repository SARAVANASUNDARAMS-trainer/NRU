```python
# Import NumPy
import numpy as np

# Display version
print(np.__version__)
```

## Vectorized Operations
Vectorization performs operations on an entire array without writing explicit loops.

```python
# Create an array
arr=np.array([10,20,30,40,50])

# Add 5 to every element
print(arr+5)

# Multiply every element by 2
print(arr*2)

# Divide every element by 10
print(arr/10)

# Square each element
print(arr**2)
```

## Python Loop vs NumPy

```python
# Python list
numbers=[10,20,30]

# Empty result list
result=[]

# Loop through each element
for x in numbers:
    # Add 5
    result.append(x+5)

# Print result
print(result)

# NumPy array
arr=np.array(numbers)

# Vectorized addition
print(arr+5)
```

## Mathematical Operations

```python
# Arrays
a=np.array([10,20,30])
b=np.array([1,2,3])

# Addition
print(a+b)

# Subtraction
print(a-b)

# Multiplication
print(a*b)

# Division
print(a/b)

# Power
print(a**b)
```

## Mathematical Functions

```python
# Sample array
arr=np.array([4,9,16,25])

# Square root
print(np.sqrt(arr))

# Exponential
print(np.exp([1,2,3]))

# Logarithm
print(np.log([1,2,3]))

# Absolute value
print(np.abs([-10,-5,8]))
```

## Statistical Operations

```python
# Marks
marks=np.array([70,80,90,60,85])

# Mean
print(np.mean(marks))

# Median
print(np.median(marks))

# Standard deviation
print(np.std(marks))

# Variance
print(np.var(marks))

# Minimum
print(np.min(marks))

# Maximum
print(np.max(marks))
```

## Broadcasting

```python
# Matrix
matrix=np.array([[1,2,3],[4,5,6]])

# Vector
vector=np.array([10,20,30])

# Broadcast addition
print(matrix+vector)
```

## Machine Learning Example

```python
# Salary feature
salary=np.array([25000,50000,75000,100000])

# Min-Max Scaling
scaled=(salary-np.min(salary))/(np.max(salary)-np.min(salary))

# Display scaled values
print(scaled)
```
